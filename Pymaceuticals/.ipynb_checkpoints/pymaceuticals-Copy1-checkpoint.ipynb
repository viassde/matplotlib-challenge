```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# File to Load
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse/Drug Data and the Clinical Trial Data
mouse = pd.read_csv(mouse_drug_data_to_load)
trial = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined = pd.merge(mouse, trial, how='outer', on='Mouse ID')

# Display the data table for preview
combined.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
gcombined = combined.groupby(['Drug', 'Timepoint'])
tmp = gcombined['Tumor Volume (mm3)'].mean()

# Convert to DataFrame
df_gcombined = pd.DataFrame({'Avg TumorVol [mm3]':tmp})

# Preview DataFrame
df_gcombined.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tmp = gcombined['Tumor Volume (mm3)'].sem()

# Convert to DataFrame
df_semgcombined = pd.DataFrame({'StdErr TumorVol [mm3]':tmp})

# Preview DataFrame
df_semgcombined.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames for Avg Tumor Vol
tmp =df_gcombined.reset_index()

tumorvol =tmp.pivot('Timepoint', 'Drug', 'Avg TumorVol [mm3]') 
tumorvol.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames for SEM of Tumor Vol
tmp =df_semgcombined.reset_index()

semtumorvol =tmp.pivot('Timepoint', 'Drug', 'StdErr TumorVol [mm3]') 
semtumorvol.head(10)
```

```python
# Generate the Plot (with Error Bars)
fig, ax = plt.subplots()

# define different markers
mrk =['^', 's', 'P', '>', '<', 'd', '*', 'o', '+', 'X']
# counter for markers
i =0

for col in tumorvol.columns:
    ax.errorbar(tumorvol.index, tumorvol[col], yerr=semtumorvol[col], fmt=mrk[i], linestyle='solid',
                alpha=0.75, label=col)
    i =i+1
    
    
ax.set_xlabel("Timepoint [days]")
ax.set_ylabel("Tumor Vol [mm3]")
ax.set_title("Tumor Response to Treatment")
ax.grid()

plt.legend(loc="best", fontsize="large", fancybox=True)

fig.set_size_inches(16, 9)
fig.savefig('Treatment.png') 
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
tmp = gcombined['Metastatic Sites'].mean()

# Convert to DataFrame
df_mgcombined = pd.DataFrame({'Avg No. Metastasis':tmp})

# Preview DataFrame
df_mgcombined.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
tmp = gcombined['Metastatic Sites'].sem()

# Convert to DataFrame
df_msemgcombined = pd.DataFrame({'StdErr No. Metastasis':tmp})

# Preview DataFrame
df_msemgcombined.head()
```

```python
# Minor Data Munging to Re-Format the Metastasis Data Frames
tmp =df_mgcombined.reset_index()
metastnbr =tmp.pivot('Timepoint', 'Drug', 'Avg No. Metastasis') 
metastnbr.head(10)
```

```python
# Minor Data Munging to Re-Format the SEM Metastasis Data Frames
tmp =df_msemgcombined.reset_index()
semmetastnbr =tmp.pivot('Timepoint', 'Drug', 'StdErr No. Metastasis') 
semmetastnbr.head(10)
```

```python
# Generate the Metastasis Plot (with Error Bars)
fig, ax = plt.subplots()

# markers 
mrk =['^', 's', 'P', '>', '<', 'd', '*', 'o', '+', 'X']
i =0

for col in metastnbr.columns:
    ax.errorbar(metastnbr.index, metastnbr[col], yerr=semmetastnbr[col], fmt=mrk[i], linestyle='solid',
                alpha=0.75, label=col)
    i=i+1

ax.set_xlabel("Timepoint [days]")
ax.set_ylabel("No. Metastatic Sites")
ax.set_title("Metastatic Response to Treatment")
ax.grid()

plt.legend(loc="best", fontsize="large", fancybox=True)

fig.set_size_inches(16, 9)
fig.savefig('Spread.png') 
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint
tmp = gcombined['Mouse ID'].count()

# Convert to DataFrame
df_rgcombined = pd.DataFrame({'No. Mice':tmp})

# Preview DataFrame
df_rgcombined.head(13)
```

```python
# Minor Data Munging to Re-Format the Data Frames
tmp =df_rgcombined.reset_index()
micenbr =tmp.pivot('Timepoint', 'Drug', 'No. Mice') 
micenbr.head(10)
```

```python
# Generate the Survival Plot (Accounting for percentages; i.e., rates)

tmp =df_rgcombined.unstack(0)
# get rate by dividing by max
tmp = tmp / tmp.max()
tmp = tmp.stack()

tmp =tmp.reset_index()
perc_micenbr =tmp.pivot('Timepoint', 'Drug', 'No. Mice') 
perc_micenbr.head(10)
```

```python
fig, ax = plt.subplots()

# differen markers
mrk =['^', 's', 'P', '>', '<', 'd', '*', 'o', '+', 'X']
# counter for markers
i =0

for col in perc_micenbr.columns:
    ax.plot(perc_micenbr.index, 100*perc_micenbr[col], marker=mrk[i], linestyle='solid',alpha=0.75, label=col)
    i =i+1
    
ax.set_xlabel("Timepoint [days]")
ax.set_ylabel("Survival Rate [%]")
ax.set_title("Survival During Treatment")
ax.grid()
plt.legend(loc="best", fontsize="large", fancybox=True)
fig.set_size_inches(16, 9)
fig.savefig('Survival.png') 
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
perc_change = {}
for col in tumorvol.columns:
    #print(col)  
    perc_change[col] =100*(tumorvol[col][45]/tumorvol[col][0] -1.0)
    
# Display the data to confirm
print(perc_change)
```

```python
# Splice the data between passing and failing drugs
tup = [(k, v) for k, v in perc_change.items()]

passtup = [(x,y) for x,y in tup  if y < 0]  # PASS
failtup =[(x,y) for x,y in tup  if y >= 0]
```

```python
# Orient widths. Add labels, tick marks, etc. 
fig, ax = plt.subplots()

rectp = ax.bar(  range(len(passtup)), [val[1] for val in passtup], color='green', align='center')
rectf = ax.bar(  range(2,10), [val[1] for val in failtup], color='red', align='center')

fintup = passtup + failtup

plt.xticks(  range(len(fintup)), [val[0] for val in fintup], fontsize="large")
plt.xticks(rotation=0)

ax.set_ylabel("Tumor Vol Change [%]")
ax.set_title("Tumor Change over 45 day treatment")
ax.grid()

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 4*np.sign(height), 
                '%d %s' % (int(height), '%'),  #'%s %d' % (name, number)
                ha='center', va='bottom', fontsize="large", color="white")
autolabel(rectp)
autolabel(rectf)

fig.set_size_inches(16, 7)
fig.savefig('Change.png')
```

## Observations about the results of the study - mainly based on visualizations generated from the study data

* Two drugs -Capomulin and Ramicane- show much better results (on average) than the rest when looking at the 3 metrics in this study - tumor volume reduction, number of metastasis, survival rates


* Capomulin and Ramicane show the following results:
    * Reduction in average tumor volume - means of 19% and 22% reduction at timepoint of 45 days, respectively. For reference, Placebo is at 51% increase, and third best drug is at 42% increase
    * Survival rates of 84% and 80% at 45 days, respectively. For reference, Placebo is at 44%, and third best drug is at 56%
    * Average number of metastasis of 1.48 and 1.25 at 45 days, respectively. For reference, Placebo is at 3.27, and third best drug is at 1.73


* The error (std error of the means) should be considered for any more precise conclusions


* Some drugs perform worse (on average) than the Placebo; for instance if we list the worst in each metric:
    * Ketapril shows 57% increase in volume (Placebo at 51%)
    * Propriva at 27% survival rate (Placebo at 44%)
    * Ketapril at 3.36 in number of metastasis (Placebo at 3.27%)
